```python
import pandas as pd     
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data_set = pd.read_csv("new_data_set.csv",encoding="cp1252") #Reading csv file that contains dataset for covid patients
```

```python
data_set.drop('Patient ID',axis=1,inplace=True)
data_set.drop(['Parainfluenza 3', 'Parainfluenza 1', 'Bordetella pertussis', 'Chlamydophila pneumoniae', 'Adenovirus'],axis=1,inplace=True)#After knowing feature performances
data_set.drop(['Mean corpuscular volume (MCV)','Hematocrit','Hemoglobin'],axis=1,inplace=True) #Removing multicollinearity
```

```python
data_set = data_set[~(data_set.iloc[:,6:].isnull().all(axis=1))]
```

```python
missing_percentage = data_set.isnull().sum()/len(data_set)*100
columns_to_drop = missing_percentage[missing_percentage>=85].index.tolist()
data_set = data_set.drop(columns_to_drop,axis=1) #Dropping columns having more than 85% null alues
```

```python
data_set = data_set[~(data_set.iloc[:,6:].isnull().all(axis=1))]
#Some rows would haven't been dropped due to those column dropped recently, hence dropping those remaining 
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 5) #Odd numbers are recommended for n_neighbors, most of the time it is 5
data_setCat = data_set.select_dtypes(include=['object']) #Extracting categorical data
for i in data_setCat.keys():
    data_setCat[i].fillna(data_setCat[i].mode()[0],inplace=True) #Replacing missing values with mode value
data_setFloat = data_set.drop(data_setCat.keys(),axis=1)
data_setFloat = data_setFloat.drop(['Patient addmited to regular ward (1=yes, 0=no)'],axis=1)#Extracting numerical values
data_setFloat[data_setFloat.keys()] = imputer.fit_transform(data_setFloat) #imputing numerical values
data_set[data_setCat.keys()] = data_setCat #Replacing missing values with imputed values
data_set[data_setFloat.keys()] = data_setFloat #Replacing missing values with imputed values
```

```python
data_set = data_set.drop('Parainfluenza 2',axis=1) 
data_set.drop_duplicates(inplace=True)#Parainfluenza 2 is constant, this was checked using .nunique() and had only 1 unique value
```

```python
data_set
```

```python
data_set = data_set.replace({'positive':1,'negative':0})
data_set = data_set.replace({'detected':1,'not_detected':0})
data_set
```

```python
corr = data_set.corr()
plt.figure(figsize=(25,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
x = data_set.drop('SARS-Cov-2 exam result',axis=1)
y = data_set[['SARS-Cov-2 exam result']]
```

```python
from imblearn.over_sampling import SMOTE       #Balancing data
smote = SMOTE()
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
import numpy as np
np.count_nonzero(y_resampled==1),np.count_nonzero(y_resampled==0)
```

```python
from sklearn.model_selection import  train_test_split  #Splitting data into train and test set
x_train, x_test, y_train, y_test= train_test_split(x_resampled.values, y_resampled.values, test_size= 0.2, random_state=0)  
```

```python
pd.DataFrame(x_train).describe()
```

```python
from sklearn.preprocessing import StandardScaler  #Standard scaling data
st_x= StandardScaler()  
x_train= st_x.fit_transform(x_train)  
x_test= st_x.transform(x_test)  
```

```python
pd.DataFrame(x_train).describe()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from sklearn.ensemble import RandomForestClassifier
#Defining the hyperparameters to search over
param_dist = {'n_estimators': randint(100, 1000),
              'max_depth': randint(1, 50),
              'min_samples_split': randint(2, 20),
              'min_samples_leaf': randint(1, 20)}

#Creating a random forest classifier object
rf = RandomForestClassifier()

#Creating a RandomizedSearchCV object
rs = RandomizedSearchCV(rf, param_distributions=param_dist, n_iter=50, cv=10)

#Fitting the RandomizedSearchCV object to the data
rs.fit(x_train, y_train)

#Printing the best hyperparameters and corresponding validation score
print('Best hyperparameters:', rs.best_params_)
print('Best validation score:', rs.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier  
#Defining classifier for Random Forest
classifier= RandomForestClassifier(n_estimators= 453,min_samples_split = 12, max_depth = 49, min_samples_leaf=1)  
#Fitting the data
classifier.fit(x_train, y_train)  
```

```python
# Get feature importances
importances = classifier.feature_importances_

# Sort features by importance in descending order
indices = importances.argsort()[::-1]

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(x.shape[1]), importances[indices])
plt.xticks(range(x.shape[1]), x.columns[indices], rotation=90)
plt.tight_layout()
```

```python
from sklearn.svm import SVC
from scipy.stats import uniform
from sklearn.model_selection import RandomizedSearchCV

# Defining the SVM classifier
svm_clf = SVC()

# Defining the hyperparameters to tune
hyperparameters = {'C': uniform(0, 10),
                   'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
                   'degree': [1,2, 3, 4, 5],
                   'gamma': ['scale', 'auto']}

# Performingn random search with 10-fold cross validation
rand_search = RandomizedSearchCV(svm_clf, hyperparameters, n_iter=50, cv=10, random_state=42, n_jobs=-1, verbose=1)

# Fitting the random search object to the training data
rand_search.fit(x_train, y_train)

# Printing the best hyperparameters and their corresponding score
print("Best Hyperparameters: ", rand_search.best_params_)
print("Best Score: ", rand_search.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_dist = {'n_neighbors': randint(1, 50),
              'p': [1, 2],
              'weights': ['uniform', 'distance']}

# Creating a KNN classifier object
knn = KNeighborsClassifier()

# Creating a RandomizedSearchCV object with cross-validation
rs = RandomizedSearchCV(estimator=knn,
                        param_distributions=param_dist,
                        cv=5,
                        n_iter=100,
                        random_state=0,
                        n_jobs=-1,
                        verbose=2)

# Fitting the RandomizedSearchCV object to the training data
rs.fit(x_train, y_train)

# Printing the best hyperparameters and corresponding score
print("Best Hyperparameters:", rs.best_params_)
print("Best Score:", rs.best_score_)
```

```python
from sklearn.linear_model import LogisticRegression
param_distributions = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': np.logspace(-3, 3, 7),
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 250, 500, 1000],
}


# Creating Logistic Regression classifier
clf = LogisticRegression()

# Creating RandomizedSearchCV object
random_search = RandomizedSearchCV(
    estimator=clf,
    param_distributions=param_distributions,
    n_iter=100,
    cv=5,
    random_state=42,
    n_jobs=-1
)

# Fitting the randomized search object to the data
random_search.fit(x_train, y_train)

# Getting best hyperparameters and best score
best_params = random_search.best_params_
best_score = random_search.best_score_
print("Best hyperparameters:", best_params)
print("Best score:", best_score)
```
